```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roll_number = 23
np.random.seed(roll_number)
```

```python
sales_data = np.random.randint(1000, 5001, size=(12, 4))
sales_data
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, index=months, columns=columns)
df
```

**Part II: Data Manipulation and Analysis (Pandas)**

```python
df.head()
df.describe()
```

```python
total_per_category = df.sum()
total_per_month = df.sum(axis=1)

total_per_category, total_per_month
```

```python
avg_growth = df.diff().mean()
avg_growth
```

```python
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100

df
```

```python
if roll_number % 2 == 0:
    df['Electronics'] = df['Electronics'] * 0.90
else:
    df['Clothing'] = df['Clothing'] * 0.85

df
```

**Part III: Visualizations (Matplotlib & Seaborn)**

```python
plt.figure(figsize=(10,6))

for col in columns:
    plt.plot(df.index, df[col], marker='o', label=col)

plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df[columns])
plt.title('Sales Distribution by Category')
plt.show()
```

***Q2: NumPy Array Operations***

```python
array = np.array([[1, -2, 3],
                  [-4, 5, -6]])
np.abs(array)
```

```python
flat = array.flatten()

np.percentile(flat, [25, 50, 75])          # Flattened
np.percentile(array, [25,50,75], axis=0)   # Column-wise
np.percentile(array, [25,50,75], axis=1)   # Row-wise
```

```python
# Flattened
np.mean(flat), np.median(flat), np.std(flat)

# Column-wise
np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0)

# Row-wise
np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1)
```

***Q3: Floor, Ceiling, Truncate, Round***

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

np.floor(a)
np.ceil(a)
np.trunc(a)
np.round(a)
```
